```python
            #------Multivariate analysis------#
# Multivariate analysis examines the relationship among three or more variables at the same time.
# Multi= Many, Variate=Variable
```

```python
# Use different plot during bivariate and multivariate analysis
# Scatterplot- Numerical and Numerical
# Bar plot-Numerical and categorial
# Box Plot-Numerical and categorial
# Dis plot-Numerical and categorial
# # Heatmap- categorial and categorial
```

```python
# Bivariate and multivarite anlysis on tips dataset
tips.info()
```

```python
# Relation between total_bill and tip column using Scatterplot 
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    alpha=0.7
)
plt.xlabel('Total_bill')
plt.ylabel('Tip')
plt.title('Relationship between Total_bill and Tip')
plt.show()
```

```python
# Multivariate analysis: Total_bill, tip and sex
plt.figure(figsize=(5,6))
sns.scatterplot(
    data=tips, 
    x='total_bill', 
    y='tip',
    hue='sex',
    )
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip by Sex and Time')
plt.legend(title='Legend')
plt.show()
```

```python
# Multivariate analysis: Total_bill, tip, time, size and sex
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=tips, 
    x='total_bill', 
    y='tip',
    hue='sex',
    style='time',
    size='size'
    )
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip by Sex and Time')
plt.legend(title='Legend')
plt.show()
```

```python
# Bivarite analysis: Numerical and categorial data

# Barplot using titanic dataset
titanic.head(3)
```

```python
# Pclass and age column
plt.figure(figsize=(3,4))
sns.barplot(data=titanic, x='Pclass', y='Age', estimator='mean', palette=['red', 'blue', ])      # Calculates mean Age for each Pclass
plt.xlabel('Passenger Class')
plt.ylabel('Average Age')
plt.title('Average Age by Passenger Class')
plt.show()
```

```python
# Pclass and fare column
plt.figure(figsize=(3,4))
sns.barplot(data=titanic, x='Pclass', y='Fare', estimator='mean', palette=['red', 'blue', 'green'])      # Calculates mean Age for each Pclass
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.title('Average Fare by Passenger Class')
plt.show()
```

```python
# Multivariate analysis on Pclass, Fare and Sex
plt.figure(figsize=(3,4))
sns.barplot(data=titanic, x='Pclass', y='Fare', hue='Sex', estimator='mean', palette=['red', 'blue', ])      # Calculates mean Age for each Pclass
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.title('Average Fare by Passenger Class')
plt.show()
```

```python
# Multivariate analysis on Pclass, age and Sex
plt.figure(figsize=(3,4))
sns.barplot(data=titanic, x='Pclass', y='Age', hue='Sex', estimator='mean', palette=['red', 'blue', ])      # Calculates mean Age for each Pclass
plt.xlabel('Passenger Class')
plt.ylabel('Average age')
plt.title('Average age by Passenger Class')
plt.show()
```

```python
# Bivariate analysis on age and survive column
sns.displot(
    data=titanic,
    x='Age',
    hue='Survived',
    kind='kde',
    fill=True
)

plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
# Calculate group of Sex who survived
titanic.groupby('Sex')['Survived'].mean()*100
```

```python
            #-------Categorial and Categorial-------# 
# For categorail and categorial use two plot:
# 1. Heatmap 
# 2. Clusterplot
```

```python
# Heatmap for categorial and categorial
# Bivariate analysis based on heatmap using sex and survived column
heatmap_data=pd.crosstab(titanic['Sex'], titanic['Survived'])
plt.figure(figsize=(6,4))
sns.heatmap(
    heatmap_data, 
    annot=True,        # show numbers
    fmt='d',           # integer format
    cmap='YlGnBu'      # color palette
)
plt.title('Survival Count by Sex')
plt.ylabel('Sex')
plt.xlabel('Survived')
plt.show()
```

```python
# Multivariate analysis
heatmap_data = pd.crosstab(titanic['Sex'], 
            [titanic['Survived'], titanic['Pclass']]) # One row and two column
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='coolwarm')
plt.title('Survival Count by Sex and Pclass')
plt.show()
```

```python
# Draw line plot on flights dataset
# A line plot shows how a variable changes along an order
# X-axis  independent variable (can be time, category, or any ordered value)
# Y-axis  dependent variable (numeric)
flight.head()
Flight_year = flight.groupby('year')['passengers'].sum().reset_index()
sns.lineplot(data=Flight_year, x='year', y='passengers')

```

```python
# Draw heatmap using flight dataset
Fp=flight.pivot_table(values='passengers', index='month', columns='year')
sns.heatmap(Fp)
```
