```python
# Univariate analysis: Analysis of a single variable.

# Uni=Single, Variate= Variable
# Independent analysis of a single column of a dataset called univariate analysis.
```

```python
# During univariate analysis first ask the datatype of a variable/column
# May be it will be numerical or categorial
```

```python
# Import labraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import dataset
Df=pd.read_csv('titanic.csv')
```

```python
# what is the size of dataset
Df.shape
```

```python
# How does the data look like
Df.head(4)
```

```python
# What is the data type of the column
Df.info()
```

```python
            #------ Know the frquency of categories of Categorial Data-----#
# For categorial data use countplot/pie chart

# CountPlot   

# Frecuency of categories of survived column
sns.countplot(x='Survived', data=Df, palette=['green', 'blue'])
plt.show
```

```python
# Plot using value_counts method
Df['Survived'].value_counts()
Df['Survived'].value_counts().plot(kind='bar')
```

```python
# For Pclass variable
sns.countplot(x='Pclass', data=Df, palette=['red', 'blue', 'yellow'])
```

```python
# For sex column
Df['Sex'].value_counts()
sns.countplot(x='Sex', data=Df, palette=['Red', 'blue'])
```

```python
# Show embarked column
sns.countplot(x='Embarked', data=Df)
```

```python
        # ----Know the distribution of each categories of categorial column through pie chart----#

# Show the pie chart of Survived column
Df['Survived'].value_counts().plot(kind='pie',  autopct='%.2f', labels=['Not Survived', 'Survived'])
```

```python
# For Pclass column
Df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# To know the distribution of numerical column use:
    # Histogram
    # Distplot
    # Boxplot
```

```python
# Histogram for age coulumn
plt.hist(Df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
# Dist plot
sns.displot(
    data=Df,
    x='Age',
    bins=30,
    kde=True
)

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```

```python
# Only Kernel Density Estimation
sns.kdeplot(Df['Age'], fill=True)
plt.show()
```

```python
# Boxplot 
# Also called box-and-whisker plot
# A box plot is used to summarize the distribution of numerical data 
# Detect outliers very clearly.

# A box plot displays five key statistics:
    # Minimum (excluding outliers)
    # 25% percentile (Q1)  first quartile
    # Median (50%)
    # 75% percentile (Q3)  third quartile
    # Maximum (excluding outliers)

# Lower whisker - Q1=25%
# Q1 - median=25%
# Median-Q3=25%
# Q3-max whisker=25%

# Lower whisker  Q1  1.5  IQR
# Upper whisker  Q3 + 1.5  IQR

# Q1 (25%)  25% of data are below this value
# Median (50%)  50% below
# Q3 (75%)  75% below
# Box = Q1 to Q3
# Contains the middle 50% of the data
```

```python
# Box plot for age column
sns.boxplot(x='Age', data=Df)
plt.xlabel('Age')
plt.title('Box Plot of Age')
plt.show()
```

```python
# Boxplot for fare
sns.boxplot(x='Fare', data=Df)
plt.title('Box Plot of Fare')
plt.show()
```

```python
# Find maximum and minimum

# Minimum
min=Df['Age'].min()
print(min)
```

```python
# Maximum
print(Df['Age'].max())
```

```python
# Find mean
print(Df['Age'].mean())
```
